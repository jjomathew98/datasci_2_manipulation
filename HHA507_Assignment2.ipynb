```python
# Identify missing values
missing_values = df.isnull().sum()
```

```python
# Handle missing values (e.g., drop rows with missing values)
df = df.dropna()
```

```python
# Verify that missing values are handled
print(df.isnull().sum())
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Remove duplicate columns (if any)
df = df.loc[:, ~df.columns.duplicated()]

# Verify that duplicates are removed
print(df.shape)
```

```python
df
```

```python
# Clean column names (e.g., remove spaces and make them lowercase)
df.columns = df.columns.str.strip().str.lower()

# Verify cleaned column names
print(df.columns)
```

# Data Transformation

```python
# Check data types of columns
df.dtypes
```

```python
# Convert 'average heart rate' column to numeric, coerce non-numeric values to NaN
df['average heart rate'] = pd.to_numeric(df['average heart rate'], errors='coerce')

# Define the heart rate threshold
threshold_high = 100

# Create a new column 'Heart_rate_status' based on 'average heart rate'
df['Heart_rate_status'] = df['average heart rate'].apply(lambda x: 'High' if x >= threshold_high else 'Normal')

```

```python
df['Heart_rate_status'] = np.where(df['average heart rate'] > 60.0, df['average heart rate'], 'below normal')
df['Heart_rate_status'] = np.where(df['average heart rate'] < 60.0, df['average heart rate'], 'normal')
df
```
